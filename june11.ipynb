```python
Q1

lambda arguments: expression

Lambda functions are typically used when you need a simple function for a short period of time, and you don't want to define a separate named function for it. They are commonly used in functional programming and in situations where you need to pass a function as an argument to another function.
Lambda functions differ from regular functions in a few key ways:

1. Lambda functions are anonymous: They don't have a name, unlike regular functions which are defined with the 'def keyword and have a name assigned to them.

2. Lambda functions are single- expression functions: They can only contain a single expression, which is evaluated and returned. In contrast, regular functions can contain multiple statements and have a return statement to specify the return value.

3. Lambda functions are typically used as function literals: They are
defined and used in the same place without assigning them to a variable.

Here's an example that demonstrates the difference between a lambda function and a regular function:

# Regular function return x ^ **2

def square(x):

# Lambda function

lambda_square = lambda x :x^ **

print(square (5))

print(lambda_square(5))

#0

#0

In this example, both the 'square

function and the lambda_square lambda function calculate the square of a number. However, the lambda function is defined inline without a name, whereas the regular function is defined using the def keyword and assigned a name

Q2
Yes, a lambda function in Python can have multiple arguments. You can define and use multiple arguments in a lambda function by separating them with commas, just like in a regular function.

Here's the syntax for defining a lambda function with multiple arguments:

lambda argument1, argument 2,
You can specify any number of arguments in the lambda function based on your requirements. The arguments are separated by commas and can be of any valid data type.

Here's an example that demonstrates a lambda function with multiple arguments:

# Lambda function with multipl multiply lambda x, y: x * y

result = multiply(5, 3)

print(result) # Output: 15

In this example, the lambda function 'multiply takes two arguments 'x' and y and returns their product. When we call the lambda function with multiply(5, 3), it multiplies 5 and 3 and returns the result, which is 15.

You can extend this concept to any number of arguments by including them in the lambda function definition and accessing them within the expression.
Q3
Lambda functions in Python are typically used in situations where you need a small, anonymous function for a short period of time. They are commonly used in conjunction with other functions that accept function objects as arguments. Lambda functions provide a concise and inline way to define these small functions without the need for explicitly naming them.

One common use case for lambda

->
One common use case for lambda functions is with higher-order functions like map(), 'filter()', and reduce(), which expect a function as one of their arguments. Lambda functions can be used to define these functions on the fly without the need for separate function definitions.

Here's an example use case where a lambda function is used with the map() function to apply a transformation to a list of numbers:

[1, 2, 3, 4, 5]

=
numbers = [1, 2, 3, 4, 5]

# Using map() with a lambda fu squared_numbers = map(lambda x

# Convert the map object to a result = list(squared_numbers)

print(result)

# Output: [1,

In this example, the lambda function `lambda x x ** 2 is used as the first argument to 'map(). It squares each element in the numbers list and returns a map object. We convert the map object to a list using `list() and store the result in the 'result' variable. The resulting list
`list() and store the result in the 'result' variable. The resulting list contains the squared numbers `[1, 4, 9, 16, 25].

Lambda functions are also commonly used in sorting and key extraction tasks, especially when you need to define custom sorting criteria or extract specific properties from objects.

Overall, lambda functions provide a concise and convenient way to define small, anonymous functions for immediate use in various contexts where function objects are required..
Q4
Lambda functions in Python offer several advantages and have some limitations compared to regular functions. Let's explore them:

Advantages of Lambda Functions:

1. Concise syntax: Lambda functions allow you to define simple functions in a compact and inline manner, without the need for a full function definition. This can make the code more concise and
straightforward operations.

2. Anonymous nature: Lambda functions are anonymous, meaning they don't require a formal name. This is advantageous when you need a function for a specific purpose and don't want to clutter your code with a separate function definition.

3. Functional programming: Lambda functions are commonly used in functional programming paradigms, where functions are treated as first-class citizens. They are particularly useful when working with higher-order functions like 'map().
'filter(), and reduce ()^ * allowing for more expressive and functional programming style.

Limitations of Lambda Functions:

1. Single expression limitation: Lambda functions are restricted to a single expression. They can't contain multiple statements or complex logic. If you need to perform more complex operations or include control flow statements, a regular function with a 'def statement is more suitable.

2. Limited readability: While lambda functions can make code more concise, excessive use of complex lambda functions can decrease
code readability. They are best suited for simple, one-line operations. If the logic becomes too complex, it is better to use a regular function with a descriptive name.

3. Lack of documentation and testability: Lambda functions lack the ability to include drawstring for documenting their purpose and usage. Additionally, they can be challenging to test independently since they are usually used in an inline manner.

4. Debugging limitations: When using lambda functions, error messages may refer to the lambda expression itself rather than the specific line within the expression
This can make debugging more challenging compared to regular functions, which have named lines and functions for error identification.

In summary, lambda functions provide a concise and anonymous way to define small functions, making them suitable for certain use cases, such as working with higher-order functions or performing simple operations. However, for more complex operations, code readability, documentation, and testability, regular functions with a full def statement are often preferred.

Q5
Yes, lambda functions in Python have access to variables defined outside of their own scope. This concept is known as "lexical scoping" or "closure." Lambda functions can access and use variables from the enclosing scope in which they are defined.
Here's an example that demonstrates the access of variables from the outer scope within a lambda function:
def outer function():

x = 10

# Lambda function accessing

inner_function = lambda y:

return inner function

closure = outer_function()

result = closure (5)

print(result)

# Output: 15

In this example, the

`outer function defines a variable

`x` with a value of 10. Inside the

outer function, a lambda function.

`inner_function is defined. The

lambda function takes an argument
Q.6
Certainly! Here's a lambda function that calculates the square of a given number:

square lambda x: * ** 2

In this lambda function, the argument

`x is squared using the exponentiation operator *. The

result of the expression overline x ^ ** 2^ - is

the square of the input overline x
the square of the input overline x ^ * .

You can use this lambda function to calculate the square of a number by calling it, like this:

result square (5) print(result) # Output: 25

In this example, the lambda function

square is called with the argument

5. It calculates the square of 5 and returns the result, which is then printed. The output will be '25', which is the square of 5.

Q7
Certainly! Here's a lambda function that finds the maximum value in a list of integers:

find_max = lambda 1st: max(1st

In this lambda function, the argument

1st represents the list of integers.

The max() function is used within the lambda function to find the maximum value from the given list.
You can use this lambda function to find the maximum value in a list by calling it, like this:

numbers = [4, 9, 2, 7, 5]

result = find_max (numbers) print(result) # Output: 9

In this example, the lambda function `find_max is called with the

Q.8
Certainly! Here's a lambda function that filters out all the even numbers from a list of integers:

filter_even lambda 1st: list

In this lambda function, the argument '1st represents the list of integers.

The filter() function is used

numbers list as the argument. It internally uses the `max() function to find the maximum value in the list. The result, which is the maximum value '9', is then printed.
In this lambda function, the argument 1st represents the list of integers.

The filter() function is used within the lambda function to create a new list that contains only the even numbers. The lambda function lambda x :*\%2==0 checks if each element 'x in the list is divisible

by 2 (i.e., an even number). If the

condition is true, the element is
included in the filtered list.

You can use this lambda function to filter out even numbers from a list by calling it, like this

numbers = [1, 2 3, 4 5 6 7 result filter_even (numbers) print (result) # Output: [2, 4

In this example, the lambda function `filter_even is called with the numbers list as the argument. It filters out the even numbers from the list using the lambda function, and the result is a new list [2, 4, 6, 8,

10] containing only the even
numbers. The result is then printed.

Q.9
Certainly! Here's a lambda function that sorts a list of strings in ascending order based on the length of each string:

sort by length = lambda 1st: s

In this lambda function, the argument '1st represents the list of strings.

The sorted() function is used

within the lambda function to sort the list based on a custom key. The lambda function `lambda x:

len(x) calculates the length of each string x in the list, and this length is used as the sorting key. The

The Sorted() function returns a new sorted list based on this custom key.

You can use this lambda function to sort a list of strings in ascending order based on their lengths by calling it, like this

strings = ['apple', 'banana', result = sort_by_length(string

print(result) # Output: ['dat
In this example, the lambda function

'sort_by_length is called with the 'strings list as the argument. It sorts the list of strings in ascending order based on their lengths using the lambda function, and the result is a new list ['date', 'apple', 'banana', 'cherry',

'elderberry']. The result is then printed.

Q10
Certainly! Here's a lambda function that takes two lists as input and returns a new list containing the common elements between the two lists:

find_common lambda lst1, 1st

In this lambda function, the arguments 1st1` and `1st2`
represent the two lists. The

'filter() function is used within the lambda function to create a new list that contains only the elements present in both lists. The lambda function lambda x:x in 1st2 checks if each element 'x' in '1st1' is present in 1st2. If the condition is true, the element is included in the filtered list.

You can use this lambda function to find the common elements between two lists by calling it, like this
list1

=

[1, 2, 3, 4,

51

list2= [4, 5, 6, 7, 81

result = find_common (list1, li print(result) # Output: [4, 5

In this example, the lambda function `find_common is called with listi and list2 as the arguments. It filters out the common elements between the two lists using the lambda function, and the result is a new list [4, 5] containing only the elements present in both lists. The result is then printed.

Q11
Certainly! Here's a recursive function in Python to calculate the factorial of a given positive integer:

def factorial (n):

if n ==0 :

return 1

else:

return n* factorial (n

In this recursive function, we check

the base case where `n` is equal to '0'
If ' n ^ * is 0, it means we have reached the smallest factorial value, which is 1. In that case, we return 1.

If 'n' is not 0, we calculate the factorial by multiplying 'n' with the factorial of n -1^ - This recursive call reduces the value of `n by 1 with each call until it reaches the base case.

You can use this recursive function to calculate the factorial of a positive integer by calling it, like this:case.

You can use this recursive function to calculate the factorial of a positive integer by calling it, like this

result = factorial(5) print(result) # Output: 120

In this example, the `factorial function is called with the argument 5. It recursively calculates the factorial of 5 as 5^ * 4 * 3 * 2 1, which equals 120. The result, 120, is then printed.

Q12
Certainly! Here's a recursive function in Python to compute the nth

Fibonacci number:

def fibonacci(n): if n <= 0 : elif n ==1 or n ==2 :

return None

return 1

else:

return fibonacci(n

1
In this recursive function, we handle three cases:

1. Base case: If 'n' is less than or equal to O, we return 'None' to handle invalid inputs.

2. Base case: If 'n' is 1 or 2, we return 1 since the first two Fibonacci numbers are both 1.

3. Recursive case: For any other value of `n`, we recursively compute the n-th Fibonacci number by summing the `(n-1)`- th and (n - 2) -th Fibonacci numbers.

You can use this recursive function to compute the nth Fibonacci number by calling it, like this:
result = fibonacci(6) print(result) # Output: 8

In this example, the 'fibonacci function is called with the argument 6. It recursively computes the 6th Fibonacci number, which is 8. The result, 8, is then printed.

Please note that recursive functions can have performance limitations for larger values of 'n', as they may result in redundant calculations. Using techniques like memoization or iterative approaches can be more efficient in such cases.

Q13
Certainly! Here's a recursive function in Python to find the sum of all the elements in a given list:

def recursive sum(1st):

if not 1st:
In this recursive function, we handle two cases:

1. Base case: If the given list 1st is empty, we return O as the sum of an empty list is 0.

2. Recursive case: For a non-empty list, we compute the sum by adding the first element of the list 1st[0]' with the sum of the remaining elements

`recursive sum (1st[1:]). This

is achieved by making a recursive call with the sublist starting from the second element onwards.

You can use this recursive function to find the sum of all the elements in a list by calling it, like this:
my_list = [1, 2, 3, 4, 5] result = recursive sum (my_list print(result) # Output: 15

In this example, the recursive sum function is called with the my_list* as the argument. It recursively computes the sum of all the elements in the list, which is 1 + 2 + 3 + 4 + 5 = 15 The result, 15, is then printed.

return 0

else:

return 1st[0] + returs


Q14
Certainly! Here's a recursive function in Python to determine whether a given string is a palindrome

def is palindrome (string):

if len(string) <= 1: return True elif string[0] != string[- return False

else:

return is palindrome (s
In this recursive function, we handle three cases:

1. Base case: If the length of the string 'string' is 0 or 1, it means the string is either empty or has a single character. In both cases, we consider it a palindrome and return 'True'.

2. Base case: If the first and last characters of the string 'string' are not equal, it means the string is not a palindrome, and we return

'False'.

3. Recursive case: For a string with more than one character and with equal first and last characters, we make a recursive call with the substring string [1:-1]. Thisrecursive call checks whether the substring is a palindrome.

You can use this recursive function to determine whether a given string is a palindrome by calling it, like this:

python

Copy code

string = "racecar"

result is_palindrome (string) print(result) # Output: True

In this example, the is_palindrome function is called with the string "racecar" as the argument. It recursively checks whether the string is a palindrome. Since the string reads the same forwards and backwards,.
is indeed a palindrome, and the result is 'True'. The result is then printed.

Please note that this recursive approach may not be the most efficient for large strings due to the overhead of recursive calls. In practice, iterative solutions are often preferred for palindrome checks
.
Q.15
Certainly! Here's a recursive function in Python to find the greatest common divisor (GCD) of two positive integers using the Euclidean algorithm:

python

def gcd (a, b):

if b ==0 :

Copy code

return

a

else:

return gcd(b, a % b)

In this recursive function, we use the Euclidean algorithm to calculate the GCD of two numbers.

The algorithm works as follows:
1. If 'b' is equal to O, it means we have found the GCD, and we return 'a'.

2. Otherwise, we make a recursive call to the 'gcd function with 'b' as the new value of 'a' and the remainder of a divided by 'b' as the new value of 'b'. This step repeats until 'b' becomes O.

You can use this recursive function to find the GCD of two positive integers by calling it, like
result = gcd (36, 48) print(result) # Output: 12

In this example, the 'gcd function is called with the two positive integers 36 and 48 as the arguments. It recursively calculates the GCD using the Euclidean algorithm. The GCD of 36 and 48 is 12, and the result, 12, is then printed.

Please note that the Euclidean algorithm is an efficient and widely used method for finding the GCD of two numbers
```
